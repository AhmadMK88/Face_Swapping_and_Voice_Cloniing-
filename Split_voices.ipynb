```python
def VideoToWav(name,path) :
  # Load the video file
  video = VideoFileClip(name)

  # Extract the audio from the video
  audio = video.audio

  # Save the audio as WAV file
  audio.write_audiofile(path+f'{name[:6]}.wav')
```

```python
def splitAudio ( filePath , savePath ) :

  global cnt
  with wave.open( filePath, 'rb') as wav_file:

    frame_rate = wav_file.getframerate()
    num_channels = wav_file.getnchannels()

    segment_frames = math.floor(5* frame_rate)
    audio_data = wav_file.readframes(segment_frames)

    while audio_data:

        segment_filename = f'segment_{cnt}.wav'
        segment_filepath = os.path.join(savePath, segment_filename)

        with wave.open(segment_filepath, 'wb') as segment_wav:

            segment_wav.setparams(wav_file.getparams())
            segment_wav.writeframes(audio_data)

        audio_data = wav_file.readframes(segment_frames)

        cnt += 1
```

```python
def CheckLength(name) :

  with wave.open(name, "r") as wav_file:

    frames = wav_file.getnframes()

    frame_rate = wav_file.getframerate()

    duration = frames / float(frame_rate)

  if duration != 5 :
    os.remove(name)
```
